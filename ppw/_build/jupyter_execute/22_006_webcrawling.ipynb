```python
import requests
import pandas as pd
```

### Penjelasan Kode Secara Keseluruhan

"""
Kode ini adalah skrip Python yang dirancang untuk mengambil data publikasi ilmiah dari **API Springer Nature**. Skrip ini bekerja dengan mengirimkan permintaan pencarian ke server API berdasarkan serangkaian kata kunci yang telah ditentukan.
"""

# ---
### Pengaturan Awal (Inisialisasi)

"""
Pada bagian ini, kode melakukan persiapan sebelum memulai proses pengambilan data.
* **Kunci API (`api_key`)**: Mengidentifikasi pengguna untuk mengakses layanan API Springer Nature.
* **Daftar Kata Kunci (`keywords`)**: Berisi topik-topik yang akan dicari, seperti `"web mining"` dan turunannya.
* **URL Endpoint API (`url`)**: Alamat server API yang akan dihubungi.
* **List Kosong (`all_records`)**: Wadah untuk menyimpan semua data publikasi yang berhasil ditemukan dari setiap kata kunci.
"""

# ---
### Proses Pengambilan Data (Looping)

"""
Proses ini menggunakan perulangan (`for`) untuk mengotomatisasi pencarian untuk setiap kata kunci dalam daftar.
* Setiap kata kunci (`kw`) dari list `keywords` akan diproses satu per satu.
* Sebuah permintaan **HTTP GET** dikirimkan ke URL API. Permintaan ini menyertakan parameter penting:
    * `q`: Kata kunci pencarian saat ini.
    * `api_key`: Kunci API Anda.
    * `p`: Jumlah hasil per halaman (diatur 20).
* Fungsi `requests.get()` bertugas untuk mengirimkan permintaan ini ke server.
"""

# ---
### Penanganan Respons (Validasi dan Ekstraksi)

"""
Setelah permintaan dikirim, kode memeriksa respons dari server untuk memastikan data berhasil diambil.
* Kode memeriksa **status respons**. Jika statusnya **`200` (OK)**, berarti permintaan berhasil.
* Data dalam format **JSON** diubah menjadi objek Python (*dictionary*) untuk diproses.
* Selanjutnya, kode melakukan perulangan (`for`) untuk mengambil setiap publikasi (`record`) dari data hasil.
* Informasi penting seperti **DOI**, **judul**, dan **abstrak** diambil. Metode `.get()` digunakan untuk menghindari error jika data tidak ditemukan.
* Setiap data yang berhasil diekstrak akan dicetak ke konsol dan disimpan ke dalam list `all_records` sebagai sebuah *dictionary*.
"""

# ---
### Penanganan Error

"""
Bagian ini memastikan skrip tidak berhenti jika terjadi kesalahan.
* Jika **status respons** dari server bukan `200` (misalnya `401` untuk otorisasi gagal atau `404` untuk halaman tidak ditemukan), kode akan mencetak pesan `Error` yang informatif.
* Pesan ini mencakup kode status dan teks respons, membantu Anda mengidentifikasi penyebab kegagalan permintaan.
"""

```python
# API Key Springer Nature
api_key = "9aa2aa513b78be3c8250c64138c84ed8"

# daftar kata kunci
keywords = ["web mining", "web usage mining", "web structure mining"]

#alamat url
url = "https://api.springernature.com/meta/v2/json"

all_records = []

for kw in keywords:
    params = {
        "q": kw,
        "api_key": api_key,
        "p": 20
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        print(f"Kata kunci: {kw} | Total hasil: {data['result'][0]['total']}\n")

        for record in data['records']:
            doi = record.get('doi', 'N/A')
            title = record.get('title', 'No title')
            abstract = record.get('abstract', 'No abstract')

            print(f"DOI: {doi}")
            print(f"Title: {title}")
            print(f"Abstract: {abstract}\n")

            all_records.append({
                "Keyword": kw,
                "DOI": doi,
                "Title": title,
                "Abstract": abstract
            })
    else:
        print("Error:", response.status_code, response.text)
```

"""
Kode ini dibawah untuk menyimpan semua data publikasi yang telah berhasil diambil dari API ke dalam sebuah file **CSV (Comma-Separated Values)**. Proses ini memastikan data yang sudah dikumpulkan dapat disimpan secara permanen dan digunakan untuk analisis lebih lanjut.
"""

# ---
### Proses Penyimpanan

"""
1.  **Mengubah List Menjadi DataFrame**:
    * `df = pd.DataFrame(all_records)`: Baris ini menggunakan pustaka **pandas** untuk mengubah `all_records` (sebuah list yang berisi kamus data) menjadi objek **DataFrame**. DataFrame adalah struktur data tabular dua dimensi, seperti tabel di *spreadsheet*, yang ideal untuk mengelola data terstruktur.

2.  **Menyimpan DataFrame ke File CSV**:
    * `df.to_csv("zain.csv", index=False, encoding="utf-8")`: Perintah ini menyimpan DataFrame ke dalam file bernama `"zain.csv"`.
        * `index=False`: Parameter ini mencegah pandas menulis indeks (nomor baris) sebagai kolom di dalam file CSV, sehingga data yang dihasilkan lebih rapi.
        * `encoding="utf-8"`: Menjamin semua karakter, termasuk karakter non-standar, disimpan dengan benar di dalam file.

3.  **Konfirmasi Penyimpanan**:
    * `print("✅ Semua hasil sudah disimpan ke zaincsv.csv")`: Setelah proses selesai, pesan konfirmasi ini dicetak untuk memberitahu Anda bahwa data telah berhasil disimpan.
"""

```python
# simpan ke CSV gabungan
df = pd.DataFrame(all_records)
df.to_csv("zain.csv", index=False, encoding="utf-8")
print("✅ Semua hasil sudah disimpan ke zaincsv.csv")
```
