Web crawling, atau sering disebut "web spider" atau "bot," adalah proses otomatis yang dilakukan oleh program komputer untuk menelusuri World Wide Web dan mengumpulkan informasi dari halaman web secara sistematis.

Secara sederhana, web crawler bekerja seperti pustakawan digital yang terus-menerus membaca dan mengumpulkan data dari miliaran halaman web yang ada di internet. Tujuan utamanya adalah untuk membuat indeks dari konten-konten tersebut agar nantinya dapat digunakan oleh mesin pencari, seperti Google, untuk menampilkan hasil yang relevan ketika pengguna melakukan pencarian.

### Cara Kerja Web Crawling

# Markdown cell untuk penjelasan bertahap
"""
Proses web crawling umumnya meliputi tahapan berikut:

1.  **Mengunjungi URL Awal:** Crawler memulai dari sebuah URL (Uniform Resource Locator) atau daftar URL yang sudah diketahui.
2.  **Membaca Konten:** Setelah sampai di sebuah halaman, crawler akan membaca seluruh konten yang ada di dalamnya, termasuk teks, gambar, video, dan metadata.
3.  **Mengikuti Tautan:** Crawler akan mengidentifikasi semua tautan (link) yang ada di halaman tersebut, baik tautan internal (ke halaman lain di situs yang sama) maupun tautan eksternal (ke situs lain).
4.  **Menambahkan ke Antrean:** Tautan-tautan yang ditemukan akan ditambahkan ke dalam antrean untuk dikunjungi selanjutnya. Proses ini berulang terus-menerus, memungkinkan crawler untuk menjelajahi jaringan situs yang luas.
5.  **Mengindeks dan Menyimpan:** Informasi yang dikumpulkan oleh crawler kemudian diindeks dan disimpan dalam database mesin pencari. Indeks ini berfungsi sebagai "peta" bagi mesin pencari untuk menemukan dan menampilkan konten yang paling relevan saat ada permintaan pencarian.
"""
